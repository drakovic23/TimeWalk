```python
import timewalk as tw
import matplotlib.pyplot as plt
from timewalk.strategy_classes import *

bt = tw.BackTester()

result = (bt.load_data("SPY", "1h").run(RsiStrategy(rsi_window=20, oversold_threshold=70, overbought_threshold=30, atr_window=20)))

result.tail()
```

```python
plt.figure(figsize=(10,6))
plt.plot(result.index, result['TotalRealPnL'], label='TotalRealPnL', linewidth=1)
plt.xlabel('Date')
plt.ylabel('Total Realized PnL')
plt.title('Total Realized P&L Over Time')
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(result.index, result['CurrentCapital'], label='Total Capital', linewidth=1, color='orange')
plt.xlabel('Date')
plt.ylabel('Total Capital')
plt.title('Capital over time')
plt.legend()
plt.grid(True)

plt.show()
```
