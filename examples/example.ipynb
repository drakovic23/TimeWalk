```python
import timewalk as tw
import matplotlib.pyplot as plt
from timewalk.strategy_classes import *

bt = tw.BackTester()

result = (bt.load_data("SPY", "1h").run(RsiStrategy(rsi_window=20, oversold_threshold=70, overbought_threshold=30, atr_window=20)))

result.tail()
```

```python
plt.figure(figsize=(10,6))
plt.plot(result.index, result['TotalRealPnL'], label='TotalRealPnL', linewidth=2)
plt.xlabel('DateTime')
plt.ylabel('TotalRealPnL')
plt.title('Total Realized P&L Over Time')
plt.legend()
plt.grid(True)  # Optional, adds a grid to the plot

plt.show()
```
